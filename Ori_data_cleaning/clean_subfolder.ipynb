```python
def stat_count(dir):
    csv_list=os.listdir(dir)
    df_store=pd.DataFrame(columns=['Categories', 'Frames','Sample rate 5','Video','Path'])
    for csv in csv_list:
        df=pd.read_csv(dir+csv)
        
        cat_list=df['Categories'].values
        
        for i,cate in enumerate(cat_list):
            if cate not in df_store['Categories'].values:
                df_store=df_store.append([{'Categories':cate}], ignore_index=True)
                index = df_store[df_store.Categories == cate].index.tolist()[0]
                df_store.loc[index,'Frames']=df.loc[i,'Frames']
                df_store.loc[index,'Sample rate 5']=df.loc[i,'Sample rate 5']
                df_store.loc[index,'Video']=df.loc[i,'Case']
                df_store.loc[index,'Path']=df.loc[i,'Path']
                    
                
            else:
                index = df_store[df_store.Categories == cate].index.tolist()[0]
                df_store.loc[index,'Frames']+=df.loc[i,'Frames']
                df_store.loc[index,'Sample rate 5']+=df.loc[i,'Sample rate 5']
                df_store.loc[index,'Video']+=df.loc[i,'Case']
                df_store.loc[index,'Path']+=df.loc[i,'Path']
                
    df_store=df_store.sort_values(by="Frames" , ascending=False)
    df_store=df_store.reset_index().drop(['index'], axis=1)
    
    df_store.to_csv('Stat.csv',index =False)
    return df_store
```

```python
def sort_csv(csv_dir):
    df=pd.read_csv(csv_dir)
    df_store=pd.DataFrame(columns=['Categories', 'Frames','Sample rate 5','Video','Path'])
    cat_list=df['Categories'].values 
    modify_class=['misssing','wufashibei','gongzuorengyuan','qitarenyuan','konpai']
    modified_class=['missing','wufashibie','gongzuorenyuan','gongzuorenyuan','kongpai']
    drop_class=['wufashibie','kongpai','cuowu']
    for i,cate in enumerate(cat_list):
        
        if cate not in df_store['Categories'].values and cate not in modify_class:
            df_store=df_store.append([{'Categories':cate}], ignore_index=True)
            index = df_store[df_store.Categories == cate ].index.tolist()[0]
            df_store.loc[index,'Frames']=df.loc[i,'Frames']
            df_store.loc[index,'Sample rate 5']=df.loc[i,'Sample rate 5']
            df_store.loc[index,'Video']=df.loc[i,'Video']
            df_store.loc[index,'Path']=df.loc[i,'Path']
    #index = df_store[df_store.Categories == cate ].index.tolist()[0]
    for a,b in zip(modify_class,modified_class):
        index_a = df[df.Categories == a ].index.tolist()[0]
        index_b = df_store[df_store.Categories == b ].index.tolist()[0]
        df_store.loc[index_b,'Frames']+=df.loc[index_a,'Frames']
        df_store.loc[index_b,'Sample rate 5']+=df.loc[index_a,'Sample rate 5']
        df_store.loc[index_b,'Video']+=df.loc[index_a,'Video']
        df_store.loc[index_b,'Path']+=df.loc[index_a,'Path']
    index_p = df_store[df_store.Categories == 'gongzuorenyuan' ].index.tolist()[0]
    df_store.loc[index_p,'Categories']='person'
    df_store=df_store.sort_values(by="Frames" , ascending=False)
    
    #df_store=df_store.reset_index().drop(['Path'], axis=1).drop(['index'], axis=1)
    df_store.reset_index()
    #df_store.drop(index=df_store[df_store['Categories'].isin(drop_class)].index[0],inplace=True)
    
    df_store.to_excel('File_path.xlsx',index =False)
    return df_store
```
